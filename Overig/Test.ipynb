```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import itertools
# plt.style.use('windesheim')

f = lambda x, a: a * x
x = np.arange(0, 11, 1)
y_true = f(x=x, a=1.5)

_ = sns.scatterplot(x=x, y=y_true)
_ = plt.xlabel('x')
_ = plt.ylabel('y')

plt.axhline(0, color='#A6A8AB', ls='--')
plt.axvline(0, color='#A6A8AB', ls='--')

plt.show()
```
